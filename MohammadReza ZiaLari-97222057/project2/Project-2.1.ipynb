```python
!pip install neurodynex
from neurodynex.brunel_model import LIF_spiking_network
from neurodynex.tools import plot_tools
import brian2 as b2
```

```python
n_Excit = 800
n_Inhib = 200
connection_probability = 0.1
w0 = 0.1 * b2.mV
g = 1
poisson_input_rate = 13. * b2.Hz
w_external = 0.1 * b2.mV
sim_time = 250. * b2.ms
```

```python
def runSim():
    rate_monitor, spike_monitor, voltage_monitor, monitored_spike_idx = LIF_spiking_network.simulate_brunel_network(N_Excit=n_Excit, 
                                                                                                                    N_Inhib=n_Inhib, 
                                                                                                                    connection_probability=connection_probability,
                                                                                                                    w0=w0,
                                                                                                                    g=g,
                                                                                                                    poisson_input_rate=poisson_input_rate, 
                                                                                                                    w_external=w_external, 
                                                                                                                    sim_time=sim_time)
    plot_tools.plot_network_activity(rate_monitor, spike_monitor, voltage_monitor, spike_train_idx_list=monitored_spike_idx, t_min=0.*b2.ms)
```

```python
runSim()
```
